#### Limpieza Fechas

```python
df_cafe['Transaction Date'] = pd.to_datetime(df_cafe['Transaction Date'], errors = 'coerce') 
```

```python
invalid_dates = df_cafe[df_cafe['Transaction Date'].isna()]
```

#### Limpieza Items

```python
df_cafe['Price Per Unit'].unique()
```

```python
df_cafe['Price Per Unit'] = pd.to_numeric(df_cafe['Price Per Unit'])
```

```python
df_cafe.loc[df_cafe['Price Per Unit'].isin(['ERROR', 'UNKNOWN']), 'Price Per Unit'] = np.nan
```

```python
df_cafe[df_cafe['Price Per Unit'].isin(['ERROR', 'UNKNOWN'])]['Price Per Unit']
```
