Limpieza de variables

```python
df.dtypes
```

```python
import unicodedata
import re
```

```python
def clean_text(text, pattern="[^a-zA-Z0-9 ]"):
    
    cleaned_text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
    cleaned_text = re.sub(pattern, " ", cleaned_text.decode("utf-8"), flags=re.UNICODE)
    cleaned_text = u' '.join(cleaned_text.lower().strip().split())
    return cleaned_text

#Creamos esta función que nos sirve para estandarizar las variables de texto
```

Limpiar variables de texto con funcion

**.map()**: Permite transformar un iterable completo (en este caso el dataframe) usando otra función

**lambda**: es una pequeña función anonima que toma multiple argumentos pero sólo una expresión. Sintaxis de lambda: *lambda argumento: expresión*

```python
df["s_nombre_del_beneficiario"] = df["s_nombre_del_beneficiario"].map(lambda x:clean_text(x.lower()).replace("\n",""))
```

```python
df["s_nombre_del_proyecto"] = df["s_nombre_del_proyecto"].map(lambda x:clean_text(x.lower()).strip().replace("\n",""))
```

```python
df["s_programa_fondo"] = df["s_programa_fondo"].map(lambda x:clean_text(x.lower()).strip().replace("\n",""))
```

```python
df = df.reset_index(drop=True)
```
