```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
fig, ax = plt.subplots()
# - fig: El lienzo completo
# - ax: Los ejes donde dibujamos
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

#Grafico de lineas
plt.plot(x, y, marker = 's', linestyle = ':', color = 'purple', linewidth = 2)
plt.title('Mi primer grafico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True, alpha = 0.3)
plt.show()
```

```python
# Grafico de barras
categorias = ['Manzanas morada', 'Platanos dominico', 'Limones Fino']
ventas = [25, 30, 15]

plt.bar(categorias, ventas, color = ['red', 'yellow', 'green'])
plt.title('Ventas de Frutas')
plt.ylabel('Cantidad vendida')
#plt.xlabel('Frutas')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Grafico de barras con seaborn
df_frutas = pd.DataFrame({'fruta_1': categorias, 'ventas_frutas_205': ventas})

plt.figure(figsize = (8 , 6))
sns.barplot(data = df_frutas, x = 'fruta_1', y = 'ventas_frutas_205', palette = 'viridis')
plt.xlabel('Tipo de Fruta', fontsize = 12)
plt.ylabel('Unidades vendidas', fontsize = 12)
#plt.xticks(rotation = 45)
#plt.grid(axis = 'x', alpha = 0.3)
plt.title('Ventas de Frutas con Seaborn')
plt.tight_layout() # Ajuste automtico
plt.show()
```
